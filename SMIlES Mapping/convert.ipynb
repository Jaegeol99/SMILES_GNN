```python
import os
import pandas as pd
from tkinter.filedialog import askdirectory, askopenfilename
from tkinter import Tk
from rdkit import Chem
from rdkit.Chem import MolToSmiles

# Tkinter 초기화
root = Tk()
root.withdraw()

# 출력 폴더, CSV 선택
output_root = askdirectory(title="Select output root directory")
if not output_root:
    exit("출력 디렉토리가 선택되지 않았습니다.")
csv_file = askopenfilename(title="Select CSV file", filetypes=[("CSV files", "*.csv"), ("All files", "*.*")])
if not csv_file:
    exit("CSV 파일이 선택되지 않았습니다.")

df = pd.read_csv(csv_file)

def numbered_mapped_smiles(smiles: str) -> str:
    """수소 생략, heavy atom만 번호 매긴 후 케큘화 없이 SMILES 생성"""
    if pd.isna(smiles) or not smiles.strip():
        return ""
    mol = Chem.MolFromSmiles(smiles)
    if mol is None:
        print(f"[Error] SMILES parse failed: {smiles}")
        return ""
    # explicit H 제거
    mol = Chem.RemoveHs(mol)
    # implicit H 완전 제거
    for atom in mol.GetAtoms():
        atom.SetNoImplicit(True)
        atom.SetNumExplicitHs(0)
    # atom map 부여
    for i, atom in enumerate(mol.GetAtoms(), start=1):
        atom.SetAtomMapNum(i)
    # sanitize / kekulize 모두 제거하고 바로 SMILES 생성
    try:
        return MolToSmiles(mol, canonical=False)
    except Exception as e:
        print(f"[Warn] MolToSmiles failed on '{smiles}': {e}")
        return ""

records = []
for idx, row in df.iterrows():
    mol_folder = str(idx + 1).zfill(5)
    for state, col in (("Mapped_SMILES", "SMILES"), ("Melting_POINTS", "Melting Point"), ("Charge", "Predicted Charge")):
        smi = row.get(col, "")
        mapped = numbered_mapped_smiles(smi)
        records.append({"Mapped_SMILES": mapped})
        print(f"[OK] {mol_folder}/{state}: {mapped}")

# 결과 저장
df_out = pd.DataFrame(records)
out_path = os.path.join(output_root, "all_numbered_mapped_smiles.csv")
df_out.to_csv(out_path, sep="\t", index=False)
print(f"[✓] Saved: {out_path}")
```
